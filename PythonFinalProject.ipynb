```python
#Data source originated via https://fueleconomy.gov/feg/download.shtml for Fuel Economy Data in 2000 
#by the US Department of Energy
#Importing the set of data showing multiple lines of vehicles
import pandas as pd
cardataframe_dir = pd.read_excel("C:/Users/Quinn's PC/Documents/00data/G6080900.xls")
#Displaying the top 100 rows of data
pd.set_option('display.max_rows', None)
cardataframe_dir.head(100)
```

```python
#Displaying the different combined MPG information for all vehicles
#take note in the row count, mean, standard deviation, minimum, maximum, percentiles
cardataframe_dir["cmb"].describe()
```

```python
#Create a scatterplot of combined MPG relative to the displacement of the engine for the vehicle
#Chose because closest to a correlating factor for the data
#In addition, we have the assumption that larger displacement decreases MPG
#Simply put, displacement is the independent variable and MPG is the dependent variable
from pandas.plotting import scatter_matrix
#We also made the size of the scatter plot larger as well
cardataframe_dir.plot(kind="scatter", x="displ", y="cmb",figsize=(20,10))
```

```python
#Show a comparison of different engine displacement types starting at the mode and decreasing in quantity
cardataframe_dir['displ'].value_counts().plot(kind = 'bar', rot=0,figsize=(20,10))
```

```python
#Importing modules of numpy, plt, and scipy for using arrays, creating a different method of scatter plot for use with linear
#regression to show the slope, and stats analysis
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
#Really quickly go over median, convert dataframe of combined mpg to string
print("Median: " + str(np.median(cardataframe_dir["cmb"])))
#Really quickly go over variance, convert dataframe of combined mpg to string
print("Variance: " + str(np.var(cardataframe_dir["cmb"])))
#Display the mode of the dataset, convert dataframe of displacement to string
print("Mode of displacement: " + str(stats.mode(cardataframe_dir["displ"])))
#Display the mode of the dataset, convert dataframe of displacement to string
print("Mode of mpg: " + str(stats.mode(cardataframe_dir["cmb"])))

```

```python
#Create Target dataframe and convert it into an array by converting matrix into array, as there are difficulties with
#using 2D matrixes vs. 1D arrays
mpgDataframe = pd.DataFrame(cardataframe_dir['cmb'])
Target = mpgDataframe.to_numpy()
Target = np.squeeze(np.asarray(Target))
#Create Feature dataframe and convert it into an array by converting matrix into array, as there are difficulties with
#using 2D matrixes vs. 1D arrays
displacementDataframe = pd.DataFrame(cardataframe_dir['displ'])
Feature = displacementDataframe.to_numpy()
Feature = np.squeeze(np.asarray(Feature))
#Create Linear Regression model using scipy by defining variables from scipy stats function linregress
slopeLinear, interceptLinear, rLinear, pLinear, std_errorLinear = stats.linregress(Feature,Target)
#Create Linear Regreeion model
def linearFunc(x):
    return slopeLinear * x + interceptLinear
#adjust size of scatterplot, label axes
plt.figure(figsize=(20,10))
plt.xlabel('Engine Displacement')
plt.ylabel('Combined miles per gallon')
#print the model and scatter plot
linearModel = list(map(linearFunc,Feature))
plt.scatter(Feature,Target)
plt.plot(Feature, linearModel)
plt.show()
```

```python
#Import r-squared module from sklearn so we can find the r-squared value of the polynomial regression
from sklearn.metrics import r2_score
#Create polynomial regression from numpy module fitting order 3
polyModel = np.poly1d(np.polyfit(Feature, Target, 3))
#print r-squared value
print("R^2: " + str(r2_score(Target,polyModel(Feature))))
#Create line for polynomial regression, scale graph, create scatter plot
polyLine = np.linspace(1, 8, 70)
plt.figure(figsize=(20,10))
plt.scatter(Feature, Target)
plt.plot(polyLine, polyModel(polyLine))
plt.show()
```

```python
#Import scikit learn for preprocessing/scaling data
from sklearn import preprocessing
#Scale displacement by 0 to 1
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))
#Scale feature
scaled_feature = minmax_scale.fit_transform(displacementDataframe.to_numpy())
#print(scaled_feature)

#Create an analysis dataframe
AnalysisDataframe = pd.DataFrame()
AnalysisDataframe['displ'] = np.squeeze(np.asarray(minmax_scale.fit_transform(displacementDataframe.to_numpy())))
AnalysisDataframe['mpg'] = mpgDataframe

#create linear model of scaled data
linearModel = list(map(linearFunc,Feature))
plt.scatter(scaled_feature,Target)
plt.plot(scaled_feature, linearModel)
plt.show()









```

```python
#Luckily 845 that can be reasonably split for training and testing for a machine learning algorithm
#To do this, we're creating two dataframes for training and testing, 
#created from our previously created scaled analysis dataframe
#To do this, we use the pandas sample feature to sample a random 80% of our original dataframe for the training dataframe
#And leftover 20% for the testing dataframe
TrainingDataframe = AnalysisDataframe.sample(frac=0.8)
TestingDataframe = AnalysisDataframe.drop(TrainingDataframe.index)

#Note: The following could have been done as a function, however with the given variables, I don't think it would have been
#as easy as what I've done

#Create Target training dataframe and convert it into an array by converting matrix into array, as there are difficulties with
#using 2D matrixes vs. 1D arrays
mpgTrainingDataframe = pd.DataFrame(TrainingDataframe['mpg'])
TrainingTarget = mpgTrainingDataframe.to_numpy()
TrainingTarget = np.squeeze(np.asarray(TrainingTarget))
#Create Feature training dataframe and convert it into an array by converting matrix into array, as there are difficulties with
#using 2D matrixes vs. 1D arrays
displacementTrainingDataframe = pd.DataFrame(TrainingDataframe['displ'])
TrainingFeature = displacementTrainingDataframe.to_numpy()
TrainingFeature = np.squeeze(np.asarray(TrainingFeature))
#create linear model of scaled training data
linearTrainingModel = list(map(linearFunc,TrainingFeature))
plt.scatter(TrainingFeature,TrainingTarget)
plt.plot(TrainingFeature, linearTrainingModel)
plt.show()

```

```python
#Create Testing training dataframe and convert it into an array by converting matrix into array, as there are difficulties with
#using 2D matrixes vs. 1D arrays
mpgTestingDataframe = pd.DataFrame(TestingDataframe['mpg'])
TestingTarget = mpgTestingDataframe.to_numpy()
TestingTarget = np.squeeze(np.asarray(TestingTarget))
#Create Testing training dataframe and convert it into an array by converting matrix into array, as there are difficulties with
#using 2D matrixes vs. 1D arrays
displacementTestingDataframe = pd.DataFrame(TestingDataframe['displ'])
TestingFeature = displacementTestingDataframe.to_numpy()
TestingFeature = np.squeeze(np.asarray(TestingFeature))
#create linear model of scaled testing data
linearTestingModel = list(map(linearFunc,TestingFeature))
plt.scatter(TestingFeature,TestingTarget)
plt.plot(TestingFeature, linearTestingModel)
plt.show()
```

```python
#Create polynomial regression from numpy module fitting order 3
polyTrainingModel = np.poly1d(np.polyfit(TrainingFeature,TrainingTarget, 3))
#print r-squared value for training set
print("R^2 for Training set: " + str(r2_score(TrainingTarget,polyTrainingModel(TrainingFeature))))
#print r-squared value for testing set
print("R^2 for Testing set: " + str(r2_score(TestingTarget,polyTrainingModel(TestingFeature))))
#print median value for training set
print("Median of mpg for training set: " + str(np.median(TrainingDataframe['mpg'])))
#print median value for testing set
print("Median of mpg for testing set: " + str(np.median(TestingDataframe['mpg'])))
#print mode value for training set
print("Mode of mpg for training set: " + str(stats.mode(TrainingDataframe['mpg'])))
#print mode value for testing set
print("Mode of mpg for testing set: " + str(stats.mode(TestingDataframe['mpg'])))
#print variance value for training set
print("Variance for training set: " + str(np.var(TrainingDataframe['mpg'])))
#print variance value for testing set
print("Variance for testing set: " + str(np.var(TestingDataframe['mpg'])))

```
